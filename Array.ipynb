```
# Transverse
def traverse_array(arr):
    """Go through the array and show each index with its value."""
    print("Here are the items in the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

# Reverse
def reverse_array(arr):
    """Show the array from last item to first."""
    print("Array in reverse order:")
    print(list(reversed(arr)))

# Sort
def sort_array(arr):
    """Show the array in order from smallest to largest."""
    print("Array sorted in ascending order:")
    print(sorted(arr))

# Insert
def insert_element(arr, index, value):
    """Add a value into the array at the given index."""
    if 0 <= index <= len(arr):
        arr.insert(index, value)
        print(f"Added {value} at index {index}")
        print("Updated array:", arr)
    else:
        print("Invalid index, please try again.")

# Modify
def modify_element(arr, index, new_value):
    """Change the value at a given index."""
    if 0 <= index < len(arr):
        arr[index] = new_value
        print(f"Changed index {index} to {new_value}")
        print("Updated array:", arr)
    else:
        print("Invalid index, please try again.")

# Delete
def delete_element(arr, index):
    """Remove the value at a given index."""
    if 0 <= index < len(arr):
        removed = arr.pop(index)
        print(f"Removed {removed} from index {index}")
        print("Updated array:", arr)
    else:
        print("Invalid index, please try again.")

def main():
    # Initialize an array
    arr = [10, 20, 30, 40, 50]
    print("Original array:", arr)

    while True:
        print("\nMenu:")
        print("1. Traverse array")
        print("2. Reverse array")
        print("3. Modify element")
        print("4. Insert element")
        print("5. Sort array")
        print("6. Delete element")
        print("7. Exit")
        choice = input("Enter your choice (1-7): ")

        if choice == '1':
            traverse_array(arr)

        elif choice == '2':
            reverse_array(arr)

        elif choice == '3':
            try:
                print("Current array:", arr)
                print("Indexes:       0   1   2   3   4")
                idx = int(input("Enter index to modify: "))
                val = int(input("Enter new value: "))
                modify_element(arr, idx, val)
            except ValueError:
                print("Please enter numbers only.")

        elif choice == '4':
            try:
                print("Current array:", arr)
                idx = int(input("Enter index to insert at: "))
                val = int(input("Enter value to insert: "))
                insert_element(arr, idx, val)
            except ValueError:
                print("Please enter numbers only.")

        elif choice == '5':
            sort_array(arr)

        elif choice == '6':
            try:
                print("Current array:", arr)
                idx = int(input("Enter index to delete: "))
                delete_element(arr, idx)
            except ValueError:
                print("Please enter a number.")

        elif choice == '7':
            print("Exiting program. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 7.")

if __name__ == "__main__":
    main()
```
